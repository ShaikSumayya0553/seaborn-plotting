```python
import pandas as pd 
import seaborn as sb #seaborn consists of inbuilt dataset
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g= sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")#in the plot we can observe  how many times tip given
plt.show() 
 
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip" )
plt.show() 
```

```python
g=sb.FacetGrid(df,col="time",hue="smoker")
g.map(plt.scatter,"total_bill","tip" )
plt.show() 
```

```python
#statistical estimation with hue
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#statistical estimation without hue
sb.barplot(x="sex",y="tip",data=df)
plt.show()
```

```python
#linear Relationship
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.pairplot(df) 
plt.show() #diagonal elements are histogram,remianing are histogram  
```

```python
#plotting selected variables
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show() 
```

```python
#adding hue color to seaborn pairplot
sb.pairplot(df,hue='size') 
plt.show() 
```

```python
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
sb.pairplot(df,diag_kind='kde')
plt.show() 
```

```python
#Adjusting plot Kind
sb.pairplot(df,kind='reg')#non diagonal  elements will be regression plot
plt.show() 
```

```python
#controlling the markers
sb.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
g=sb.pairplot(df,hue='day')
g.fig.suptitle("pairplot of Tips Dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
